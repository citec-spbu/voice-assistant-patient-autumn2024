# Checking pyttsx3

```python
import pyttsx3
engine = pyttsx3.init()
engine.setProperty('rate', 120) 
engine.setProperty('volume', 1.2)
engine.say('Я могу говорить! С ужасным акцентом')
engine.runAndWait()
```

## pyttsx3 available voices (wihout ru)

```python
voices = engine.getProperty('voices')
for voice in voices:
    print('-----------------------')
    print(f'Имя: {voice.name}-')
    print(f'ID: {voice.id}')
    print(f'Язык: {voice.languages}')
    print(f'Пол: {voice.gender}') 
    print(f'Возраст: {voice.age}')
```

# speakerpy

```python
from speakerpy.lib_speak import Speaker
from speakerpy.lib_sl_text import SeleroText

speaker = Speaker(model_id="v3_1_ru", language="ru", speaker="xenia", device="cpu")
text = "Добрый день! Вы записывались на прием?"
```

## Stream audio

```python
speaker.speak(text=text, sample_rate=48000, speed=1.1)
```

# MP3

```python
speaker.to_mp3(text=text, name_text="Текст", sample_rate=48000, audio_dir="./mp3", speed=1.2)
```
